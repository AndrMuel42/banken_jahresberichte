```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Häufigkeit_UBS_Begriffe.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the import
df.head(10)
```

```python
# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Aktionär'], marker='o', linestyle='-', color='b')
plt.plot(df['Year'], df['Dividende'], marker='o', linestyle='-', color='r')
plt.plot(df['Year'], df['Reputation'], marker='o', linestyle='-', color='g')

plt.title('Häufigkeit der Begriffe über die Jahre')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
